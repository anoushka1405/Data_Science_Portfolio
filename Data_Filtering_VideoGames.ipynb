```python
#Filtering games by genre
df[df['Genre']=='Sports']
```

```python
#Filtering Games by Sales in North America
df[df['NA_Sales'] > 1]
```

```python
#Filter Games by Platform and Year
df.loc[(df['Platform']=='PS2')& (df['Year']==2005)]
```

```python
#Filter Games by Global Sales Range
df[(df['Global_Sales']>=20)&(df['Global_Sales']<=30)]
```

```python
#Filter Games by Publisher and Genre
df[(df['Publisher']=='Nintendo')&(df['Genre']=='Action')]
```

```python
# Filter Games by Multiple Genres
df.loc[df['Genre'].isin(['Action','Sports'])]
```

```python
#Select Games by Multiple Criteria

df.loc[(df['Genre'].isin(['Action','Sports']))& 
                              (df['Platform']=='PS2')&
                               ((df['Global_Sales']>=5))&(df['Global_Sales']<=30)]
```

```python
# Filter Games by Sales

df.loc[(df['Global_Sales']>=5)&(df['Global_Sales']<=30)]
```

```python
#Filter Games by Platform and Critic Score

df[(df['Platform']=='PS2')&(df['Global_Sales']>5)]
```

```python
# Filtering Rows and Columns
df.loc[(df['Year']>2000)&(df['Global_Sales']>5),
                           ['Name','Platform','Year','Global_Sales']]
```

```python
# Filter the dataset to include games with a platform of either PS4 or NES
df.loc[df['Platform'].isin(['PS4','NES']),
                            ['Name','NA_Sales','EU_Sales']]
```

```python
# Get the top 10 games from all regions
df.nlargest(10, 'Global_Sales')[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
```

```python
#Calculate the average sales in North America
df.loc[df['Genre'] == 'Action', 'NA_Sales'].mean()
```

```python
#Determine the average sales in Europe for the game published by Electronic Arts
df.loc[df['Publisher']=='Electronic Arts','EU_Sales'].mean()
```

```python
# Find all the games released between 2005 and 2010 (both inclusive) with multiple condition
df[((df['Year']>=2005)&(df['Year']<=2010))&
                      (df['Genre']=='Action')&
                      (df['Global_Sales']>5)]
```

```python
# Find the top 10 games with multiple filters
df[(df['Genre'] == 'Shooter') & (df['Year'] > 2010)].nlargest(10, 'NA_Sales')[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
```
